```python
# Import the choice function of the random module
# https://stackoverflow.com/questions/306400/how-to-randomly-select-an-item-from-a-list
from random import choice

# Assign to a list the 3 possible options: 'stone', 'paper' or 'scissors'.
gestures = ["rock","paper","scissors"]

# Assign a variable to the maximum number of games: 1, 3, 5, etc ...
max_rounds = 9

# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games
rounds_to_win = int(max_rounds/2) + 1


# Define a function that randomly returns one of the 3 options.
# This will correspond to the play of the machine. Totally random.
def choose_cpu():
    cpu_choice = choice(gestures)
    return cpu_choice

# Define a function that asks your choice: 'stone', 'paper' or 'scissors'
# you should only allow one of the 3 options. This is defensive programming.
# If it is not stone, paper or scissors keep asking until it is.
def choose_player():
    choice = input("Please select rock, paper or scissors")
    if(choice == "rock") or (choice == "paper") or (choice == "scissors"):
        return choice
    else:
        print("Sorry I couldn't get that. Please try again")
        choice == None

# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins
def winner(cpu,player):
    if cpu == player:
        return 0
    elif cpu == "rock":
        if player == "scissors":
            return 1
        else: #player == paper
            return 2
    elif cpu == "paper":
        if player == "scissors":
            return 2
        else: #player == rock
            return 1
    else: #cpu == scissors
        if player == "rock":
            return 2
        else: #player == paper
            return 1


    
    
# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated
def outcome_of_round(cpu_choice, player_choice):
    #cpu_choice = choose_cpu()
    #player_choice = choose_player()
    
    global cpu_score
    global player_score
    
    print("The CPU has chosen", cpu_choice)
    print("The player has chosen", player_choice)
    
    if winner(cpu_choice, player_choice) == 0:
        #tie do nothing
        print("This round was a tie!")
    
    elif winner(cpu_choice, player_choice) == 1:
        #cpu wins round
        print("CPU wins this round!")
        cpu_score +=1


    else:
        #player wins
        print("Player wins this round!")
        player_score +=1

        
    
# Create two variables that accumulate the wins of each participant
cpu_score = 0
player_score = 0

# Create a loop that iterates while no player reaches the minimum of wins
# necessary to win. Inside the loop solves the play of the
# machine and ask the player's. Compare them and update the value of the variables
# that accumulate the wins of each participant.
rounds_played = 0
round_counter = 1
print("we will play", max_rounds,"rounds")
print("rounds to win:", rounds_to_win)
while(cpu_score < rounds_to_win) and (player_score < rounds_to_win) and (rounds_played <max_rounds):
    
    print("Starting round", round_counter, "\n")
    cpu_choice = choose_cpu()
    player_choice = choose_player()
    while player_choice == None:
        player_choice = choose_player()
    outcome_of_round(cpu_choice,player_choice)
    rounds_played += 1
    round_counter += 1
    
# Print by console the winner of the game based on who has more accumulated wins
if cpu_score > player_score:
    print("\n\nCPU wins!!!")
elif cpu_score < player_score:
    print("\n\nPlayer wins!!!")
else:
    print("\n\nIts a tie!!!")  
```

Improvements:

```python
# Import the choice function of the random module
from random import choice

# Define a function that asks for an odd number on the keyboard, until it is not valid
# will keep asking
def request_rounds():
    rounds = int(input("How many rounds would you like to play (must be an odd number)"))
    if rounds%2 == 1:
        print("Great we will play with a limit of", rounds, "rounds")
        rounds_to_win = int(rounds/2) + 1
        print("Rounds required to win:", rounds_to_win)
        return rounds
    else:
        print("Sorry that number is not valid. Please try again!\n")
        request_rounds()

# Assign a list of 5 possible options.
new_gestures = ["rock","paper","scissors","lizard","spock"]

# Assign a variable to the maximum number of games: 1, 3, 5, etc ...
# This time the previously defined function is used
rounds = request_rounds()

# Assign a variable to the number of games a player must win to win.
# Preferably the value will be based on the number of maximum games
rounds_to_win = int(rounds/2) + 1

# Define a function that randomly returns one of the 5 options.
# This will correspond to the play of the machine. Totally random.
def new_choose_cpu():
    cpu_choice = choice(new_gestures)
    return cpu_choice

# Define a function that asks your choice between 5
# you should only allow one of the 5 options. This is defensive programming.
# If it is not valid, keep asking until it is valid.
def new_choose_player():
    choice = input("Please select rock, paper, scissors, lizard or spock")
    if(choice == "rock") or (choice == "paper") or (choice == "scissors") or (choice == "lizard") or (choice == "spock"):
        return choice
    else:
        print("Sorry I couldn't get that. Please try again")
        choice == None

# Define a function that resolves a combat.
# Returns 0 if there is a tie, 1 if the machine wins, 2 if the human player wins
# Now there are more options
def new_winner(cpu,player):
    if cpu == player:
        return 0
    elif cpu == "rock":
        if player == "scissors":
            return 1
        elif player == "paper":
            return 2
        elif player == "spock":
            return 2
        else: #lizard
            return 1
        
    elif cpu == "paper":
        if player == "scissors":
            return 2
        elif player == "rock":
            return 1
        elif player == "spock":
            return 1
        else: #lizard
            return 2
        
    elif cpu == "spock":
        if player == "scissors":
            return 1
        elif player == "rock":
            return 1
        elif player == "paper":
            return 2
        else: #lizard
            return 2
        
    elif cpu == "lizard":
        if player == "scissors":
            return 2
        elif player == "rock":
            return 2
        elif player == "paper":
            return 1
        else: #spock
            return 1      
        
    else: #cpu == scissors
        if player == "rock":
            return 2
        elif player == "spock":
            return 2
        elif player == "lizard":
            return 1
        else: #player == paper
            return 1


    
# Define a function that shows the choice of each player and the state of the game
# This function should be used every time accumulated points are updated
def new_outcome_of_round(cpu_choice, player_choice):
    
    global cpu_score
    global player_score
    
    print("The CPU has chosen", cpu_choice)
    print("The player has chosen", player_choice)
    
    if winner(cpu_choice, player_choice) == 0:
        #tie do nothing
        print("This round was a tie!")
    
    elif winner(cpu_choice, player_choice) == 1:
        #cpu wins round
        print("CPU wins this round!")
        cpu_score +=1


    else:
        #player wins
        print("Player wins this round!")
        player_score +=1
    
# Create two variables that accumulate the wins of each participant
cpu_score = 0
player_score = 0

# Create a loop that iterates while no player reaches the minimum of wins
# necessary to win. Inside the loop solves the play of the
# machine and ask the player's. Compare them and update the value of the variables
# that accumulate the wins of each participant.
rounds_played = 0
round_counter = 1

while(cpu_score < rounds_to_win) and (player_score < rounds_to_win) and (rounds_played <rounds):
    
    print("Starting round", round_counter, "\n")
    cpu_choice = new_choose_cpu()
    player_choice = new_choose_player()
    while player_choice == None:
        player_choice = new_choose_player()
    new_outcome_of_round(cpu_choice,player_choice)
    rounds_played += 1
    round_counter += 1
    
# Print by console the winner of the game based on who has more accumulated wins
if cpu_score > player_score:
    print("\n\nCPU wins!!!")
elif cpu_score < player_score:
    print("\n\nPlayer wins!!!")
else:
    print("\n\nIts a tie!!!")
```
